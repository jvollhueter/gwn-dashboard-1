# Explorative Analyse der Grundwasserneubildung
Das Notebook demonstriert einen Zeitraumvergleich mit synthetischen Jahreswerten.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'jahr': range(1961, 2021),
    'gwn_mm_a': [155 + ((year * 17) % 41) - 20 for year in range(1961, 2021)],
})
data.head()
```

```python
referenz = data.query('1961 <= jahr <= 1990')['gwn_mm_a'].mean()
vergleich = data.query('1991 <= jahr <= 2020')['gwn_mm_a'].mean()
ergebnis = pd.DataFrame({
    'periode': ['1961–1990', '1991–2020'],
    'mittelwert_mm_a': [referenz, vergleich],
})
ergebnis
```

```python
ax = data.plot(x='jahr', y='gwn_mm_a', figsize=(11, 4), legend=False)
ax.set_ylabel('Grundwasserneubildung [mm/a]')
ax.grid(True, alpha=0.3)
plt.show()
ergebnis.to_csv('gwn_zeitraumvergleich.csv', index=False)
```
